```python
from __future__ import print_function
import cv2
import numpy as np
from PIL import Image, ImageEnhance, ImageFilter
import pytesseract


lower_red = np.array([0, 0, 0])
upper_red = np.array([20, 255, 150])

vc = cv2.VideoCapture(0)

img_counter = 0

def image_process(img):
    contrast_enhancer = ImageEnhance.Contrast(img)
    pil_enhanced_image = contrast_enhancer.enhance(2)
    enhanced_image = np.asarray(pil_enhanced_image)
    r1, g1, b1 = cv2.split(enhanced_image)
    enhanced_image = cv2.merge([b1, g1, r1])
    brightness_enhancer = ImageEnhance.Brightness(img)
    pil_enhanced_image_2 = brightness_enhancer.enhance(2)  # type: Image
    enhanced_image_2 = np.asarray(pil_enhanced_image_2)
    r2, g2, b2 = cv2.split(enhanced_image_2)
    enhanced_image_2 = cv2.merge([b2, g2, r2])

    cv2.imshow('Enhanced Image', enhanced_image_2)
    return enhanced_image


while True:

    ret, frame = vc.read()
    #frame = cv2.flip(frame, 1)       # horizontal flip, change second parameter to 0 (vertical flip)
    frame = frame[0:1000, 160:480]   # croping (first parameter are the y axis, second is for the x axis)

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)   # convert color format from bgr to gray scale

    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)     # convert color format from bgr to hsv
    hsv = cv2.GaussianBlur(hsv, (9,9), 0)     # blurring/smoothing image to redure noise
                                              # cv.2.blur() can be used too

    mask = cv2.inRange(hsv, lower_red, upper_red) # pixel in the image with the color in the range will turn white, while the rest turn black

    mask, contours, hierarchy = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)  # second parameter is the method to detect contour, the third parameter is the medhod to draw the contour
                                                                                                    # can try other medhod too
    cv2.imshow('name_of _window', frame)
    k = cv2.waitKey(1)

    if k%256 == 27:
        # ESC pressed
        print("Escape hit, closing...")
        break
    elif k%256 == 32:
        # SPACE pressed
        img_name = "opencv_frame_{}.png".format(img_counter)
        cv2.imwrite(img_name, frame)
        print("{} written!".format(img_name))
        path = "C:/Users/MdssakthI/.PyCharmCE2018.2/config/scratches/opencv_frame_"+str(img_counter)+".png"
        img_counter += 1
        print(path)
        img = Image.open(path)
        img_1=image_process(img)
        #gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # convert color format from bgr to gray scale
        #hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)  # convert color format from bgr to hsv
        #img.save('temp.png')
        #pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files (x86)\Tesseract-OCR\tesseract.exe"
        #os.remove('temp.png')


```
